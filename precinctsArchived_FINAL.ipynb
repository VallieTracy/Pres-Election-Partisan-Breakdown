```python
# Import dependencies
import pandas as pd
import numpy as np
import geopandas as gp
#testing
```

```python
# Set file path
csvpath = "Resources/2016_precincts_newVersion.csv"

# Read csv and preview dataframe
prec_df = pd.read_csv(csvpath, encoding="ISO-8859-1")
pd.set_option('display.max_columns', None)
prec_df.head()
```

```python
# Pull out the columns we want
prec_df = prec_df[['VTDID', 'PCTNAME', 'PCTCODE', 'MCDNAME', 'COUNTYNAME', 'COUNTYCODE',
                   'USPRSR', 'USPRSDFL', 'USPRSTOTAL']][:-1]
prec_df
```

```python
# Change column names around candidates
prec_df = prec_df.rename(columns = {'USPRSR': 'Trump', 'USPRSDFL': 'Clinton',
                                    'USPRSTOTAL': 'Total Votes'})
prec_df
```

```python
# Add percentage columns
prec_df['Trump %'] = ((prec_df['Trump'] / prec_df['Total Votes']) * 100).round(2)
prec_df['Clinton %'] = ((prec_df['Clinton'] / prec_df['Total Votes']) * 100).round(2)
prec_df
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import geopandas as gp
from geopandas import GeoDataFrame
```

```python
# load shp file extracted from zipfile
prec_shp = gp.read_file("Resources/general_election_results_by_precinct_2016.shp" )
prec_shp
```

```python
# Change VTDID dtype for the merge below, otherwise will get an error
prec_shp['VTDID'] = prec_shp['VTDID'].astype('float64')
prec_shp.dtypes
```

```python
# Take out columns
prec_shp = prec_shp[['VTDID', 'PCTNAME', 'PCTCODE', 'MCDNAME', 'COUNTYNAME', 'COUNTYCODE', 'Shape_Leng',
                     'Shape_Area', 'geometry']]
prec_shp
```

```python
# Merge prec_df and prec_shp on VTDID
df_shp_merge = prec_df.merge(prec_shp, on = 'VTDID')
df_shp_merge
```

```python
# Pull out the desired columns
df_shp_merge = df_shp_merge[['VTDID', 'PCTNAME_x', 'PCTCODE_x', 'MCDNAME_x', 'COUNTYNAME_x', 'COUNTYCODE_x',
                       'Trump', 'Clinton', 'Total Votes', 'Trump %', 'Clinton %', 'Shape_Leng', 'Shape_Area', 
                       'geometry']]
df_shp_merge.head()
```

```python
# Rename the columns
df_shp_merge = df_shp_merge.rename(columns={'PCTNAME_x': 'PCTNAME', 'PCTCODE_x': 'PCTCODE',
                                      'MCDNAME_x': 'MCDNAME', 'COUNTYNAME_x': 'COUNTYNAME',
                                      'COUNTYCODE_x': 'COUNTYCODE'})
df_shp_merge.head()
```

```python
df_shp_merge['geometry'] = df_shp_merge['geometry'].astype(str)
df_shp_merge.dtypes
```

```python
df_shp_merge.to_csv('Resources/precincts_geodata.csv')
```
